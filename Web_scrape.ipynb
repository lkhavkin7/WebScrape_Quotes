```python
#imports
import requests
from bs4 import BeautifulSoup
```

```python
#url example, quote website
url = 'http://quotes.toscrape.com'

# sends a get request to the server
response = requests.get(url)
#print a status code whether you are allowed to use or a rejection( 200 is good)
print(response.status_code)
```

```python
# this wraps the reponse and captures it ina structered format?
soup = BeautifulSoup(response.text, 'html.parser')
#print(soup)
# literally just the soup object
# but used the prettify method to make it look better when printed
print(soup.prettify())
```

```python
quotes = soup.find_all("span", class_="text")
authors = soup.find_all("small", class_="author")

for q, a in zip(quotes, authors):
    print(f"{q.get_text()} — {a.get_text()}")
```

```python
import csv

with open("quotes.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Quote", "Author"])  # header row

    page = 1
    while True and page != 6:
        url = f"http://quotes.toscrape.com/page/{page}/"
        response = requests.get(url)
        if response.status_code != 200:
            break

        soup = BeautifulSoup(response.text, "html.parser")
        quotes = soup.find_all("span", class_="text")
        authors = soup.find_all("small", class_="author")

        for q, a in zip(quotes, authors):
            writer.writerow([q.get_text(), a.get_text()])

        page += 1
```
